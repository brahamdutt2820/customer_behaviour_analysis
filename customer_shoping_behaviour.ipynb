```python
df.columns= df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_")
df=df.rename(columns={"purchase_amount_(usd)":"purchased_amount"})
```

```python
df.columns
```

```python
lables=["young","adult","middle_aged","senior"]
df["age_group"]=pd.qcut(df["age"],q=4,labels=lables)
```

```python
df[["age","age_group"]].sample(10)
```

```python
df.groupby("frequency_of_purchases")["customer_id"].count()
```

```python
frequency_maping={
"Annually":"365",
"Bi-Weekly":"14",
"Every 3 Months":"90",
"Fortnightly":"14",
"Monthly":"30",
"Quarterly":"90",
"Weekly":"7"
}

df["purchase_frequency_days"]=df["frequency_of_purchases"].map(frequency_maping)
```

```python
df.groupby("purchase_frequency_days")["customer_id"].count()
```
